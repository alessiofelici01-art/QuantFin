# Valuation of Restricted Stock Units under IFRS 2 Framework

## 1. Introduction

Restricted Stock Units (RSUs) represent equity-settled share-based compensation instruments requiring specific accounting treatment under IFRS 2 – Share-based Payment. This document analyzes the regulatory framework and applicable valuation methodologies.

## 2. Characteristics of RSUs

RSUs are conditional rights to acquire ordinary shares, characterized by:
- Predetermined vesting period
- Performance and/or service conditions
- Absence of discretionary exercise by the beneficiary
- Settlement through share delivery

## 3. Accounting Classification

### 3.1 Equity-Settled Transactions
According to IFRS 2.2, RSUs constitute equity-settled transactions when:
- The entity receives services as consideration
- The obligation is settled by delivery of the entity's own shares
- No financial obligation exists for the entity

### 3.2 Classification Implications
The equity-settled qualification determines:
- Recognition at fair value on grant date
- Absence of subsequent remeasurement
- Recognition in equity

## 4. Valuation Methodologies

### 4.1 Fair Value at Grant Date
Fair value must be determined in accordance with IFRS 2.10-2.12:

**Market price**: When available, constitutes the primary benchmark.

**Valuation models**: In the absence of quoted prices, apply:
- Binomial model
- Black-Scholes-Merton model (adapted for RSUs)
- Monte Carlo simulations

### 4.2 RSU-Specific Considerations
Unlike stock options, RSUs do not present:
- Time value component implicitly positive
- Material time decay
- Strike price

Fair value therefore coincides with the underlying share price, adjusted for:

$$FV_{RSU} = S_0 \times P(vesting) \times D_{dividends}$$

Where:
- $S_0$ = share price at grant date
- $P(vesting)$ = probability of vesting
- $D_{dividends}$ = discount factor for foregone dividends

## 5. Accounting Recognition

### 5.1 Expense Allocation

$$Expense_{Period} = \frac{FV_{RSU} \times N_{RSU}}{Vesting_{Period}}$$

The amount is recognized in profit or loss with a corresponding credit to equity.

### 5.2 Treatment of Service Conditions
Service conditions affect the estimated probability of vesting. IFRS 2.19 requires adjustments when vesting is not probable.

### 5.3 Performance Conditions
According to IFRS 2.15-2.17, performance conditions must be reflected in the fair value estimate, not in immediate recognition.

## 6. Critical Considerations

- **Historical volatility**: Affects valuation only in case of embedded optionality
- **Dividend timing**: Relevant in present value calculation
- **Discount rate**: May be implicit in the valuation model
- **Dilution effect**: Must consider impact on earnings per share
## 7. Accounting Impact Calculation

### 7.1 Impact on Financial Statements

The accounting treatment of RSUs affects three primary financial statement components:

**Profit or Loss Impact**:
$$P\&L_{Impact} = \frac{FV_{RSU} \times N_{RSU} \times P(vesting)}{Vesting_{Period}}$$

**Equity Impact**:
- Credit to Share-based Payment Reserve (within equity)
- Reflects the obligation to deliver shares at vesting

**Earnings Per Share Impact**:
$$EPS_{Dilution} = \frac{Net\_Income}{Weighted\_Avg\_Shares + Dilutive\_RSUs}$$

RSUs are treated as potentially dilutive instruments and included in diluted EPS calculations using the treasury stock method.

### 7.2 Numerical Example

**Grant Date Parameters**:
- Grant Date: January 1, 2024
- Number of RSUs granted: 10,000
- Fair value at grant date (share price): €50.00
- Vesting period: 3 years (linear vesting)
- Probability of vesting: 95% (service condition)
- Expected dividend per share over vesting period: €1.50

**Calculation**:

1. **Fair value adjustment for dividends**:
    $$FV_{RSU} = €50.00 - €1.50 = €48.50$$

2. **Total compensation cost**:
    $$Total\_Cost = €48.50 \times 10,000 \times 95\% = €460,750$$

3. **Annual expense (2024-2026)**:
    $$Annual\_Expense = \frac{€460,750}{3} = €153,583.33$$

4. **Journal entry (each year)**:
    - Debit: Share-based Payment Expense €153,583.33
    - Credit: Share-based Payment Reserve €153,583.33

5. **EPS impact (assuming 500,000 weighted average shares outstanding)**:
    - Dilutive shares (year 1): 10,000 × (1/3) = 3,333 shares
    - Diluted shares: 503,333

## 8. Conclusions

RSU valuation under IFRS 2 requires rigorous methodology, focused on fair value determination at grant date, with systematic allocation throughout the vesting period, in accordance with IFRS 2.10 et seq. Proper accounting recognition ensures transparent financial reporting and accurate EPS presentation.



```python
import pandas as pd
from datetime import datetime, timedelta



def calculate_rsu_accounting_impact(
    grant_date,
    num_rsu,
    fair_value_grant_date,
    vesting_period_years,
    probability_vesting,
    expected_dividend,
    reporting_periods_per_year=4
):
    
    # 1. Calcolo fair value adjusted per dividendi
    fv_rsu = fair_value_grant_date - expected_dividend
    
    # 2. Calcolo costo totale di compensazione
    total_compensation_cost = fv_rsu * num_rsu * probability_vesting
    
    # 3. Calcolo spesa per periodo
    total_periods = vesting_period_years * reporting_periods_per_year
    expense_per_period = total_compensation_cost / total_periods
    
    # 4. Generazione calendario trimestrale
    periods = []
    cumulative_expense = 0
    current_date = grant_date
    
    for period_num in range(1, total_periods + 1):
        cumulative_expense += expense_per_period
        
        # Calcolo delle azioni dilutive (treasury stock method)
        dilutive_shares = num_rsu * (period_num / total_periods)
        
        periods.append({
            'Periodo': f'Q{((period_num - 1) % 4) + 1} {grant_date.year + (period_num - 1) // 4}',
            'Data Fine Trimestre': current_date + timedelta(days=90),
            'Spesa Trimestrale (€)': round(expense_per_period, 2),
            'Spesa Cumulata (€)': round(cumulative_expense, 2),
            'Azioni Dilutive (Stima)': round(dilutive_shares, 0),
            'Percentuale Vesting': f"{round((period_num / total_periods) * 100, 1)}%"
        })
        
        current_date += timedelta(days=90)
    
    # Creazione DataFrame
    df_impact = pd.DataFrame(periods)
    
    # Calcolo sommario
    summary = {
        'Parametro': [
            'Fair Value (grant date)',
            'Fair Value adjusted (dividendi)',
            'Numero RSU',
            'Probabilità Vesting',
            'Costo Totale Compensazione',
            'Periodo Vesting',
            'Spesa Annuale Media'
        ],
        'Valore': [
            f'€{fair_value_grant_date:.2f}',
            f'€{fv_rsu:.2f}',
            f'{num_rsu:,.0f}',
            f'{probability_vesting*100:.1f}%',
            f'€{total_compensation_cost:,.2f}',
            f'{vesting_period_years} anni',
            f'€{total_compensation_cost / vesting_period_years:,.2f}'
        ]
    }
    
    df_summary = pd.DataFrame(summary)
    
    return df_impact, df_summary

# Esempio di utilizzo con i dati del documento
grant_date = datetime(2024, 1, 1)
df_impact, df_summary = calculate_rsu_accounting_impact(
    grant_date=grant_date,
    num_rsu=10000,
    fair_value_grant_date=50.00,
    vesting_period_years=3,
    probability_vesting=0.95,
    expected_dividend=1.50
)

print("=" * 80)
print("PARAMETRI RSU")
print("=" * 80)
print(df_summary.to_string(index=False))
print("\n")
print("=" * 80)
print("IMPATTI CONTABILI TRIMESTRALI")
print("=" * 80)
print(df_impact.to_string(index=False))
```
