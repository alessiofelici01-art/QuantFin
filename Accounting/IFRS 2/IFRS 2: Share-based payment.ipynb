# IFRS 2: Share-based Payment

## 1. Introduction

IFRS 2 establishes the accounting and disclosure requirements for share-based payment transactions. This standard applies when an entity receives goods or services in exchange for equity instruments or when an entity incurs liabilities based on share prices or other equity instrument values. The primary objective is to ensure that entities recognize the effects of share-based payments in their financial statements and provide relevant information to users.

## 2. Scope and Applicability

IFRS 2 applies to all public and private entities that undertake share-based payment arrangements with employees or other parties. The scope encompasses:

- **Employee share plans**: stock options, restricted stock units (RSUs), and performance shares
- **Cash-settled share-based payments**: liabilities settled in cash based on equity instrument values
- **Share-based payment transactions with non-employees**: payments for goods or services received

## 3. Key Definitions

### Equity-Settled Share-Based Payments
Transactions where the entity receives goods or services and incurs an obligation to transfer equity instruments to the counterparty.

### Cash-Settled Share-Based Payments
Transactions resulting in liabilities to be settled by transferring cash or other assets, with amounts determined by reference to share prices.

### Vesting Conditions
Conditions that must be satisfied for the counterparty to become entitled to the equity instruments or cash payment, including:
- **Service conditions**: continuous employment or service provision
- **Performance conditions**: achieving specified performance metrics

## 4. Measurement Principles

### Fair Value Determination
- Measured at the grant date for equity-settled transactions
- Measured at each reporting date for cash-settled transactions
- Estimated using valuation models (e.g., Black-Scholes, binomial models) for options
- Considers vesting conditions and contractual terms

### Accounting Treatment

**Equity-Settled Arrangements:**
```
Dr. Employee Expense / Asset
    Cr. Equity (Share-based Payment Reserve)
```

**Cash-Settled Arrangements:**
```
Dr. Employee Expense / Liability
    Cr. Liability
    (adjusted at each reporting date)
```

## 5. Recognition and Expense Timing

The entity recognizes goods or services received as an expense, with a corresponding credit to equity or a liability. The expense is recognized over the vesting period on a straight-line basis, unless the vesting schedule suggests a different pattern. This ensures that the cost of share-based payments is matched with the period in which the employee provides services.

## 6. Disclosure Requirements

Entities must disclose information enabling users to understand the nature and extent of share-based payment arrangements, including:

- Description of each scheme
- Number and weighted average exercise prices of options
- Range of exercise prices
- Weighted average remaining contractual life
- Expense recognized during the period
- Assumptions used in fair value measurements

## 7. Practical Applications and Considerations

Share-based payments represent a significant component of compensation in modern organizations. Proper application of IFRS 2 ensures transparent financial reporting and provides investors with essential information regarding employee incentive structures and their impact on financial performance.
### Principal Instruments: RSUs, PSUs, and PSOs

- **RSU (Restricted Stock Units)**  
    Units of shares assigned to the employee, subject to vesting restrictions. After the vesting period, the beneficiary receives the shares or their cash equivalent. These are equity-settled instruments used to retain employees.

- **PSU (Performance Share Units)**  
    Units of shares conditioned on both the service period and the achievement of performance objectives (e.g., EBITDA, revenue growth). If conditions are satisfied, the employee receives the shares or their cash equivalent.

- **PSO (Performance Stock Options)**  
    Options on shares exercisable subject to specific company or individual performance conditions. They allow the purchase of shares at a preset price only if performance targets are met and, sometimes, if the share price exceeds a certain threshold.
