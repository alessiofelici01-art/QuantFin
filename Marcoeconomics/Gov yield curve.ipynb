```python
%%sql eurostat_yc <<
WITH yc_api AS (
    SELECT
        CAST(TIME_PERIOD AS DATE) AS date,
        geo AS country,
        maturities AS maturity,
        CAST(OBS_VALUE AS DOUBLE) AS yield
    FROM read_csv_auto(
        'https://ec.europa.eu/eurostat/api/dissemination/sdmx/2.1/data/irt_euryld_d/.Y1+Y2+Y5+Y10+Y20+Y30.DE+IT+FR/?format=SDMX-CSV&startPeriod=2026-01-01'
    )
    WHERE OBS_VALUE IS NOT NULL
)
SELECT *
FROM yc_api
ORDER BY date DESC, country ASC, maturity ASC;
```
