# Ordinary Least Squares (OLS) Regression

## Definition
Ordinary Least Squares is a linear regression method that estimates the coefficients of a linear model by minimizing the sum of squared residuals between observed and predicted values.

## Mathematical Formulation

### Model Specification
The linear model is specified as:
$$y = X\beta + \varepsilon$$

where:
- $y$ is an $n \times 1$ vector of observations
- $X$ is an $n \times k$ matrix of regressors (with first column of ones for intercept)
- $\beta$ is a $k \times 1$ vector of coefficients
- $\varepsilon$ is an $n \times 1$ vector of disturbances

### Estimation
The OLS estimator minimizes the sum of squared residuals:
$$\hat{\beta} = \arg\min_{\beta} \sum_{i=1}^{n} (y_i - x_i'\beta)^2$$

The closed-form solution is:
$$\hat{\beta} = (X'X)^{-1}X'y$$

## Key Assumptions (Gauss-Markov)
1. **Linearity**: The relationship between variables is linear
2. **Exogeneity**: $E[\varepsilon | X] = 0$ (regressors uncorrelated with errors)
3. **No Multicollinearity**: $X$ has full column rank
4. **Homoscedasticity**: $\text{Var}(\varepsilon | X) = \sigma^2 I_n$
5. **No Autocorrelation**: $E[\varepsilon_i \varepsilon_j | X] = 0$ for $i \neq j$

## Properties
Under the Gauss-Markov assumptions, the OLS estimator is **BLUE** (Best Linear Unbiased Estimator):
- **Unbiased**: $E[\hat{\beta}] = \beta$
- **Minimum Variance**: Among all linear unbiased estimators

## Variance-Covariance Matrix
$$\text{Var}(\hat{\beta}) = \sigma^2 (X'X)^{-1}$$

where $\sigma^2$ is estimated by: $\hat{\sigma}^2 = \frac{\text{SSR}}{n-k}$ (SSR = sum of squared residuals)

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
n = 100
X = np.random.randn(n, 1)
beta_true = np.array([2, 3])
y = beta_true[0] + beta_true[1] * X.ravel() + np.random.randn(n) * 0.5

# OLS estimation with scikit-learn
model = LinearRegression()
model.fit(X, y)

# Get predictions and residuals
y_pred = model.predict(X)
residuals = y - y_pred

# Calculate statistics
n, k = X.shape
k += 1  # include intercept
SSR = np.sum(residuals**2)
r2 = r2_score(y, y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred))

# Summary table
summary_df = pd.DataFrame({
    'Coefficient': ['Intercept', 'Slope'],
    'Estimate': [model.intercept_, model.coef_[0]]
})
print("OLS Regression Summary (scikit-learn):")
print(summary_df)
print(f"R² Score: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")

# Plot
plt.figure(figsize=(10, 5))
plt.scatter(X, y, alpha=0.6, label='Data')
plt.plot(X, y_pred, 'r-', linewidth=2, label='Fitted line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('OLS Regression (scikit-learn)')
plt.grid(True, alpha=0.3)
plt.show()
```
