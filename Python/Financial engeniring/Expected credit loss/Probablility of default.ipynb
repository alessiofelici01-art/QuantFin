# Expected Credit Loss (ECL)

## Definition
Expected credit loss is a forward-looking quantitative measure of credit risk for a financial asset or portfolio. It represents the present value of expected future credit losses, conditional on information available at the measurement date, and is used for provisioning and risk management.

## Key Components
- **Probability of Default (PD)**: the likelihood that the obligor will default within a specified horizon.
- **Exposure at Default (EAD)**: the expected outstanding amount at the time of default, including relevant undrawn commitments.
- **Loss Given Default (LGD)**: the proportion of exposure that is not recoverable after default, net of collateral and recovery costs.
- **Discount Factor (DF)**: the present-value adjustment, typically based on the asset’s effective interest rate or a risk-free curve with credit spread adjustments.

## Quantitative Formulation
The lifetime expected credit loss is given by:
$$
\text{ECL} = \sum_{t=1}^{T} \left[ \text{PD}_t \times \text{EAD}_t \times \text{LGD}_t \times \text{DF}_t \right]
$$
where:
- $t$ indexes future time periods up to the lifetime horizon $T$,
- $\text{PD}_t$ is the default probability in period $t$,
- $\text{EAD}_t$ is the exposure at default in period $t$,
- $\text{LGD}_t$ is the loss given default in period $t$,
- $\text{DF}_t$ is the discount factor to present value in period $t$.

## Asset Valuation
Expected credit loss is integral to the valuation of financial assets under IFRS 9. For assets measured at amortised cost, the carrying amount equals the gross carrying amount less the allowance for expected credit losses:
$$
\text{Carrying Amount} = \text{Gross Carrying Amount} - \text{ECL}
$$
In a present-value framework:
$$
\text{Value} = \sum_{t=1}^{T} \left[ \text{CF}_t - \text{ECL}_t \right] \times \text{DF}_t
$$
where $\text{CF}_t$ are contractual cash flows and \(\text{ECL}_t\) are expected losses in period $t$.

For instruments measured at fair value, ECL influences the credit risk adjustment in the valuation model, modifying the discount margin or the price that a market participant would require. The asset valuation therefore reflects both expected cash flows and the probability-weighted credit impairment.

## Theoretical Context
Under IFRS 9, ECL is measured as an unbiased, probability-weighted estimate of credit losses, incorporating multiple economic scenarios. The measurement can reflect either objective or risk-neutral probabilities depending on accounting and regulatory frameworks.

## Practical Requirements
A rigorous quantitative treatment of ECL requires:
- consistent segmentation of exposures by credit risk characteristics,
- calibration of PD and LGD models using relevant historical and forward-looking data,
- incorporation of macroeconomic scenarios to adjust default and recovery expectations,
- clear definition of the lifetime horizon for default risk measurement.

## Implications
Expected credit loss aggregates anticipated default events, exposures, and severities, discounted to present value, and provides the quantitative basis for provisioning, credit risk management, regulatory reporting, and the valuation of credit-sensitive assets.
